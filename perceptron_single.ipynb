```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.utils import np_utils
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
img_width  = X_train.shape[1]
img_height = X_train.shape[2]

print(img_width, img_height)
```

```python
is_five_train = y_train == 5
is_five_test = y_test == 5
labels = ["Not Five", "Is Five"]
```

```python
print(X_train.shape, is_five_train.shape)
```

```python
model=Sequential()
model.add(Flatten(input_shape=(img_width,img_height)))
model.add(Dense(1, activation='relu'))  # add activation func
model.compile(loss='mse', optimizer='adam',
                metrics=['accuracy'])

history = model.fit(X_train, is_five_train, epochs=50, validation_data=(X_test, is_five_test))
```

```python
def display():
  plt.plot(history.history['accuracy'])
  plt.plot(history.history['val_accuracy'])
  plt.title('model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('epoch')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()
  # summarize history for loss
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('model loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()

```
